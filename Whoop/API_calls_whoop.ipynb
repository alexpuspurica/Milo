This code is used to read from the user credentials file

```python
import os
from dotenv import load_dotenv

load_dotenv()

WhoopUser = os.getenv("USERNAMEinENV")
WhoopPass = os.getenv("PASSWORDinENV")
```

This code is used to create a WhoopClient

```python
from datetime import date, timedelta

#get yesterday's and today's date for API calls for recovery
start_date = date.today() - timedelta(1)
end_date = date.today()


from whoop import WhoopClient

with WhoopClient(WhoopUser, WhoopPass) as client:
    #INSERT API CALLS
    WhoopProf = client.get_profile()
    WhoopBodMeas = client.get_body_measurement()
    WhoopRecov = client.get_recovery_collection(start_date, end_date)
    pass
```
